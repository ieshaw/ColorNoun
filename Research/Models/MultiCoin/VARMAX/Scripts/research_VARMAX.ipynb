```python
ticker_by_mc = []
for dict in cmc_json[1:]:
    ticker_by_mc.append(dict['symbol'])
```

```python
ret_list = []
```

```python
for i in range(20):
    
    coin = ticker_by_mc[i]

    freq = 'hour'
    coin_df = pd.DataFrame((bit2.get_candles('BTC-{}'.format(coin), tick_interval= freq)['result']))
    
    #if no market, then skip this coin
    if len(coin_df) < 1:
        continue

    coin_df.index = coin_df['T']

    x_df = pd.DataFrame(index= coin_df['T'])

    x_df['spread_{}'.format(coin)] = coin_df['H'] - coin_df['L']

    x_df[['volumefrom_{}'.format(coin) ,'volumeto_{}'.format(coin)]] = coin_df[['BV' ,'V']]

    normal_len = 5

    x_df = (x_df - x_df.rolling(window= normal_len, min_periods= normal_len).mean())\
           /x_df.rolling(window=normal_len, min_periods=normal_len).std()

    ret_df = coin_df['C'].pct_change()

    x_df['return_{}'.format(coin)] = ret_df

    x_df = x_df.dropna()
    
    #check if any inf entries
    
    if x_df.max().max() == np.inf:
        continue

    if i == 0:

        X_df = x_df

        i = 1

    else:

        X_df = X_df.join(x_df, how = 'inner')
        
    VAR_model = VAR(X_df.values)

    results = VAR_model.fit(1)
    
    predictions = np.zeros((X_df.shape[0],X_df.shape[1]))

    col_num = len(X_df.columns)

    for i in range(0,X_df.shape[0]):
        predictions[i] = results.forecast(X_df.values[i,:].reshape(1, col_num), steps=1)

    pred_df = pd.DataFrame(data=predictions, index=X_df.index, columns=X_df.columns)

    # Drop columns that are not returns
    for column in pred_df.columns:

        if column.split('_')[0] != 'return':

            pred_df.drop(column, 1, inplace=True)
            
    ret_df = X_df.copy()
    for column in ret_df.columns:

        if column.split('_')[0] != 'return':

            ret_df.drop(column, 1, inplace=True)
            
    bt = backtest(true_df = ret_df, pred_df= pred_df)
    
    bt.run_backtest()
    
    ret_list.append(bt.strat_return)
    
```
